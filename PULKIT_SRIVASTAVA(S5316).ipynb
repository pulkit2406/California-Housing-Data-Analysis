#1-What is the average median income of the data set and check the distribution of data using appropriate plots. Please explain the distribution of the plot.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the California dataset

df = pd.read_excel('/content/housing+(1).xlsx')

# Calculate the average median income
average_median_income = df['median_income'].mean()

# Print the average median income
print(f'Average Median Income: ${average_median_income:.2f}')

# Create a histogram to visualize the distribution of median income
plt.figure(figsize=(8, 6))
sns.histplot(df['median_income'], bins=30, kde=True)  # Create a histogram with a kernel density estimate (KDE)
plt.xlabel('Median Income')
plt.ylabel('Frequency')
plt.title('Distribution of Median Income')
plt.show()

# Explaining the distribution of the plot:
# The histogram above shows the distribution of median income in the California dataset.
# - The x-axis represents different income ranges.
# - The y-axis represents the frequency (how many housing units fall into each income range).
# - The bars in the histogram represent the number of housing units in each income range.
# - The smooth line (KDE) provides an estimate of the probability density function of the data, showing the overall shape of the distribution.

# Observations:
#-In the histogram, the distribution of median income is right-skewed, meaning that most housing units have lower median incomes, and there are fewer units with higher median incomes. This is can be observ from the longer tail on the right side of the distribution.
#-The KDE plot shows the probability density estimation, highlighting the mode (the peak) of the distribution.
# In this specific dataset, we can see the distribution of median income among housing units in California. The histogram helps us understand whether the income is normally distributed, skewed, or exhibits other patterns.
```

#6. Write a programming construct (create a user defined function) to calculate the median value of the data set wherever required.

```python
def calculate_median(data):
    """
    Calculate the median of a list of numerical data.

    Parameters:
    - data (list or pandas Series): The dataset for which the median needs to be calculated.

    Returns:
    - median (float): The calculated median value.
    """
    sorted_data = sorted(data)  # Sort the data in ascending order
    n = len(sorted_data)

    # Check if the number of data points is even or odd
    if n % 2 == 0:
        # If the number of data points is even, the median is the average of the middle two values.
        middle1 = sorted_data[n // 2 - 1]
        middle2 = sorted_data[n // 2]
        median = (middle1 + middle2) / 2
    else:
        # If the number of data points is odd, the median is the middle value.
        median = sorted_data[n // 2]

    return median


```

#10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.



```python
# Define a function to categorize bedroom size
def categorize_bedroom_size(total_bedrooms):
    if total_bedrooms <= 10:
        return 'small'
    elif 11 <= total_bedrooms < 1000:
        return 'medium'
    else:
        return 'large'

# Apply the categorize_bedroom_size function to create the new column
df['total_bedroom_size'] = df['total_bedrooms'].apply(categorize_bedroom_size)

# Verify the new column
print(df[['total_bedrooms', 'total_bedroom_size']])

# This categorization allows for better understanding and analysis of the dataset, particularly in assessing the distribution of housing units based on their bedroom sizes.
```
